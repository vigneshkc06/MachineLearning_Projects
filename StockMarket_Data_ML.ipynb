# Imports

```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import seaborn as snb
%matplotlib inline
```

```python
from datetime import datetime
```

```python
start = datetime(2006,1,1)
end = datetime(2016,1,1)
```

# Getting Bank Stock Data

```python
def getData(ticker):
    return data.DataReader(ticker, 'yahoo', start, end)
```

```python
BAC = getData('BAC')
BAC.head()
```

```python
CITI = getData('C')
CITI.head()
```

```python
GS = getData('GS')
GS.head()
```

```python
JPM = getData('JPM')
JPM.head()
```

```python
MS = getData('MS')
MS.head()
```

```python
WFC = getData('WFC')
WFC.head()
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
bank_stocks = pd.concat([BAC,CITI,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
bank_stocks.head()
```

```python
 bank_stocks.xs('Close',axis=1,level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+' Return']= bank_stocks[tick]['Close'].pct_change()
```

```python
monthwise = BAC[BAC.index.year == 2008].reset_index()[['Date','Close']]
monthwise['30 Day Avg'] = monthwise.rolling(window=30).mean()
monthwise[5:]
```

##  rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
monthwise[monthwise['30 Day Avg'] > 0].groupby('Date').mean().plot(figsize=(12,6))
```
